# **No Divedend BSE**

```python
from scipy.stats import norm
import math

def bse_option_value(S0,E,r,T,sigma,option_type='call'):
    
    d1 = ( ( (r + 0.5 * sigma ** 2) * T ) + math.log(S0 / E) ) / (sigma * T ** 0.5)
    d2 = d1 - sigma * T ** 0.5

    if option_type == 'call':
        option_price = S0 * norm.cdf(d1) - E * math.exp(-r * T) * norm.cdf(d2)
    
    elif option_type == 'put':
        option_price = E * math.exp(-r * T) * norm.cdf(-d2) - S0 * norm.cdf(-d1)

    else:
        raise ValueError("option_type must be 'call' or 'put'")
    
    return option_price
```

```python
def bin_paramaters(sigma,T, N):
    dt= T / N

    u = math.exp(sigma * dt ** 0.5)
    d = math.exp(-sigma * dt ** 0.5)
    return {
        "up": u, "down": d
    }
```

```python
S0, E, r, T, option_type = 40, 32, 0.1, 0.5, "put"
sigma = 0.54202

result = bse_option_value(S0, E, r, T, sigma, option_type)
print(f"The {option_type} option price under BSE is: {result:.4f}")

print(bin_paramaters(sigma, T, 30))
```
